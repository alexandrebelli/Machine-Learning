# Titanic Machine Learning -- Tutorial Kaggle

```python
import pandas as pd

test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
train.head()
```

First, I would like to understand which features are useful to predict who will survive. For instance, I do not believe the Embarked harbor played a role. As a first test, lets drop the columns that I consider unuseful.

```python
train_simple = train.drop(['Name','Ticket','Cabin','Embarked'],axis = 1)
test_simple = test.drop(['Name','Ticket','Cabin','Embarked'],axis = 1)

train_simple.head()
```

```python
train_simple["Sex"] = train_simple["Sex"].map({"female": 1, "male": 2}).dropna()
test_simple["Sex"] = test_simple["Sex"].map({"female": 1, "male": 2}).dropna()
train_simple.fillna(0, inplace=True)
test_simple.fillna(0, inplace=True)
test_simple.tail()
```

```python
from sklearn.ensemble import RandomForestClassifier

train_simple_prepared = train_simple.drop(['Survived'], axis = 1)

forest_reg = RandomForestClassifier(class_weight = "balanced")

forest_reg.fit(train_simple_prepared, train_simple['Survived'])

import numpy as np
predictions = forest_reg.predict(test_simple)

predictions
```

```python
test_simple['Survived'] = predictions
submission = test_simple.drop(['Pclass','Sex','Age','Parch','Fare','SibSp'],axis=1)
submission.head()
```

```python
submission.to_csv('submission.csv',index=False)
```
