# Earthquakes Vizualization

https://www.kaggle.com/usgs/earthquake-database

I am doing simple earthquake visualization and trying Cartopy. I am also playing a bit with ML.

```python
import pandas as pd
dataset = pd.read_csv('database.csv')
dataset.head()
```

```python
dataset['Date'] =  pd.to_datetime(dataset['Date'], utc=True)
dataset['day'] = dataset['Date'].dt.day
dataset['month'] = dataset['Date'].dt.month
dataset['year'] = dataset['Date'].dt.year
dataset.tail()
```

```python
dataset_clean = dataset.drop(['Magnitude Seismic Stations','Horizontal Distance','Horizontal Error','Depth Error','Magnitude Error', 'Depth Seismic Stations','Azimuthal Gap','Root Mean Square','ID','Status', 'Location Source','Magnitude Source'],axis =1)
dataset2016 = dataset_clean.where(dataset_clean['year'] == 2016).dropna()

dataset2016.head()
```

Ok, now that I dropped the NaN values and kept the value for 2016, I would like to visualize the earthquakes this year. I should have a preview of the tectonic plate boundaries, where most of the earthquakes usually occur.

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
ax = plt.axes(projection=ccrs.Mollweide())
ax.stock_img()



plt.plot(dataset2016['Longitude'], dataset2016['Latitude'],
         color='blue', linewidth=2, marker='.',ls ='',ms=10,
         transform=ccrs.Geodetic(),
         )


plt.title('2016 Earthquakes',fontsize=30)


plt.show()
```

```python
dataset_bombs = dataset_clean.where(dataset_clean['Type'] == 'Nuclear Explosion').dropna()
dataset_bombs.head()
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
ax = plt.axes(projection=ccrs.Mollweide())
ax.stock_img()



plt.plot(dataset_bombs['Longitude'], dataset_bombs['Latitude'],
         color='blue', linewidth=2, marker='.',ls ='',ms=10,
         transform=ccrs.Geodetic(),
         )


plt.title('Nuclear Explosions After 1965',fontsize=30)


plt.show()
```

# Can we predict the earthquake latitudes for 2016?

I want to see if I can predict something from this dataset. I know it is unlikely that one's predict earthquakes, at least with our current knowledge, but I am curious to see how well a simple random forest does the job. So, let's try to predict the latitude knowing the longitude, The longitude knowing the latitude and the date.

```python
train = dataset.drop(['Time','Type','Date','Magnitude Type','Source','Magnitude Seismic Stations','Horizontal Distance','Horizontal Error','Depth Error','Magnitude Error', 'Depth Seismic Stations','Azimuthal Gap','Root Mean Square','ID','Status', 'Location Source','Magnitude Source'],axis =1)
dataset2016_train = train.where(dataset_clean['year'] == 2016).dropna()

test = dataset2016_train.drop(['Latitude'],axis=1)
test.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

train_prepared = train.drop(['Latitude'], axis = 1)

forest_reg = RandomForestRegressor()

forest_reg.fit(train_prepared, train['Latitude'])

predictions = forest_reg.predict(test)

```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
ax = plt.axes(projection=ccrs.Mollweide())
ax.stock_img()



plt.plot(dataset2016['Longitude'], dataset2016['Latitude'],
         color='blue', linewidth=2, marker='.',ls ='',ms=10,
         transform=ccrs.Geodetic(),label = 'True Latitude'
         )
plt.plot(dataset2016['Longitude'], predictions,
         color='red', linewidth=2, marker='.',ls ='',ms=10,
         transform=ccrs.Geodetic(),label=' Predicted Latitude'
         )

plt.title('2016 Earthquakes',fontsize=30)

plt.legend(fontsize=15, loc='upper right')
plt.show()
```

# Can we predict the earthquake longitudes for 2016?

```python
train2 = dataset.drop(['Time','Type','Date','Magnitude Type','Source','Magnitude Seismic Stations','Horizontal Distance','Horizontal Error','Depth Error','Magnitude Error', 'Depth Seismic Stations','Azimuthal Gap','Root Mean Square','ID','Status', 'Location Source','Magnitude Source'],axis =1)
dataset2016_train2 = train2.where(dataset_clean['year'] == 2016).dropna()

test2 = dataset2016_train2.drop(['Longitude'],axis=1)
test2.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

train_prepared2 = train2.drop(['Longitude'], axis = 1)

forest_reg2 = RandomForestRegressor()

forest_reg2.fit(train_prepared2, train2['Longitude'])

import numpy as np
predictions2 = forest_reg2.predict(test2)
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
ax = plt.axes(projection=ccrs.Mollweide())
ax.stock_img()



plt.plot(dataset2016['Longitude'], dataset2016['Latitude'],
         color='blue', linewidth=2, marker='.',ls ='',ms=10,
         transform=ccrs.Geodetic(),label = 'True Longitude'
         )
plt.plot(predictions2, dataset2016['Latitude'],
         color='red', linewidth=2, marker='.',ls ='',ms=10,
         transform=ccrs.Geodetic(),label=' Predicted Longitude'
         )

plt.title('2016 Earthquakes',fontsize=30)

plt.legend(fontsize=15, loc='upper right')
plt.show()
```

# Can we predict the day for 2016?

```python
train3 = dataset.drop(['Time','Type','Date','Magnitude Type','Source','Magnitude Seismic Stations','Horizontal Distance','Horizontal Error','Depth Error','Magnitude Error', 'Depth Seismic Stations','Azimuthal Gap','Root Mean Square','ID','Status', 'Location Source','Magnitude Source'],axis =1)
dataset2016_train3 = train3.where(dataset_clean['year'] == 2016).dropna()

test3 = dataset2016_train3.drop(['day'],axis=1)
test3.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

train_prepared3 = train3.drop(['day'], axis = 1)

forest_reg3 = RandomForestRegressor(n_estimators=100)

forest_reg3.fit(train_prepared3, train3['day'])

import numpy as np
predictions3 = np.round(forest_reg3.predict(test3))
from sklearn.metrics import accuracy_score

accuracy_score(dataset2016_train3['day'],predictions3)
```
